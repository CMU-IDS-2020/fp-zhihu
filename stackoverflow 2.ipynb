```python
def get_answered_questions_tags_by_user(user_ids):
    user_ids = [str(u) for u in user_ids]
    string = '('+', '.join((user_ids))+')'
    questions_query = f"SELECT a.id, a.body, a.owner_user_id AS uid, q.id AS qid, q.tags AS qtags, a.tags AS atags \
                FROM `bigquery-public-data.stackoverflow.posts_questions` AS q \
                INNER JOIN `bigquery-public-data.stackoverflow.posts_answers` AS a \
                    ON q.id = a.parent_id \
                WHERE a.owner_user_id IN {string}"
    # Set up the query (cancel the query if it would use too much of 
    # your quota, with the limit set to 1 GB)
    safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**11)
    questions_query_job = client.query(questions_query, job_config=safe_config)

    # API request - run the query, and return a pandas DataFrame
    questions_results = questions_query_job.to_dataframe()
    questions_results.to_csv('answers.csv')
    return questions_results

# get_answered_questions_tags_by_user(users)
```

```python
def get_user_qtag_matrix():
    user_num = len(users)
    tags_num = len(tags)
    
    shape = (user_num, tags_num)
    rows, cols = user_num, tags_num
    data = []

    for i in range(user_num):
        user_id = users['id'].iloc[i]
        questions_tags = questions[questions['owner_user_id']==user_id]['tags'].to_list()
        tags_dict = collections.Counter()
        for q in questions_tags:
            q = str(q)
            for ele in q.split('|'):
                tags_dict[ele]+=1
    
        for t in tags_dict:
            data.append([user_id, t, tags_dict[t]])

            
    df = pd.DataFrame(data, columns = ['uid', 'tag', 'value']) 
    
#     df = df.tag.str.get_dummies().groupby(df['uid']).max().reset_index()
    df = df.pivot_table(index = "uid", columns = "tag", values = 'value')
    df = df.fillna(0)
    
    df.to_csv('user_qtag.csv')
    print(df)
    
    return df

# user_qtag_matrix = get_user_qtag_matrix()
```

```python
user_qtag_matrix = pd.read_csv('user_qtag.csv')
user_atag_matrix = pd.read_csv('user_atag.csv')
```

```python
def get_correlation(user_tag_matrix, users, eps=1e-14):
    users_id = users.loc[:, ['id']]
    matrix = users_id.merge(user_tag_matrix, left_on = "id", right_on = "uid", how = "left")
    matrix = matrix.fillna(0)
    matrix = matrix.iloc[:, 2:].to_numpy()
    
    norm = np.linalg.norm(matrix,axis=1)
    norm = np.outer(norm, norm)+eps
    correlation = np.dot(matrix, matrix.T)/norm
    
    return correlation
correlation = get_correlation(user_atag_matrix, users)
```

```python
def get_similar_users(correlation, users, user_id, top=5):
    u_index = users.index[users['id'] == user_id].tolist()[0]
    arr = correlation[u_index]
    similar_index = arr.argsort()[-top:][::-1]
    similar_ids = users.loc[list(similar_index), ['id', 'display_name']]
    return similar_ids

similar = get_similar_users(correlation, users, 3460, top=5)
```
